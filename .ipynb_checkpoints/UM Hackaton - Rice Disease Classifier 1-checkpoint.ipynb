## 2.0 - Dataset

---
<br>

From the directory, we shall have 1600 training images and ~500 validation images.


Load the images using `keras.preprocessing`

### Training Dataset

```python
IMG_SIZE = (300,300)
```

```python
# Training Dataset for Brown Spot vs Healthy
brown_spot_dir = 'datasets/RiceDiseaseDataset/train/BrownSpot_vs_Healthy'

train_brownspot_ds = tf.keras.preprocessing.image_dataset_from_directory(
    brown_spot_dir,
    seed=123,
    image_size=IMG_SIZE,
    label_mode='binary'
)
```

```python
# Training Dataset for Hispa vs Healthy
hispa_dir = 'datasets/RiceDiseaseDataset/train/Hispa_vs_Healthy'

train_hispa_ds = tf.keras.preprocessing.image_dataset_from_directory(
    hispa_dir,
    seed=123,
    image_size=IMG_SIZE,
    label_mode='binary'
)
```

```python
# Training Dataset for Leaf Blast vs Healthy
leafblast_dir = 'datasets/RiceDiseaseDataset/train/LeafBlast_vs_Healthy'

train_leafblast_ds = tf.keras.preprocessing.image_dataset_from_directory(
    leafblast_dir,
    seed=123,
    image_size=IMG_SIZE,
    label_mode='binary'
)
```

### Validation Dataset

```python
# Validation Dataset for Brown Spot vs Healthy
brown_spot_dir = 'datasets/RiceDiseaseDataset/validation/BrownSpot_vs_Healthy'

validate_brownspot_ds = tf.keras.preprocessing.image_dataset_from_directory(
    brown_spot_dir,
    seed=123,
    image_size=IMG_SIZE,
    label_mode='binary'
)
```

```python
# Validation Dataset for Hispa vs Healthy
hispa_dir = 'datasets/RiceDiseaseDataset/validation/Hispa_vs_Healthy'

validate_hispa_ds = tf.keras.preprocessing.image_dataset_from_directory(
    hispa_dir,
    seed=123,
    image_size=IMG_SIZE,
    label_mode='binary'
)
```

```python
# Validation Dataset for Leaf Blast vs Healthy
leafblast_dir = 'datasets/RiceDiseaseDataset/validation/LeafBlast_vs_Healthy'

validate_leafblast_ds = tf.keras.preprocessing.image_dataset_from_directory(
    leafblast_dir,
    seed=123,
    image_size=IMG_SIZE,
    label_mode='binary'
)
```

### Class Names

```python
brownspot_class_names = train_brownspot_ds.class_names
hispa_class_names = train_hispa_ds.class_names
leafblast_class_names = train_leafblast_ds.class_names

print(brownspot_class_names)
print(hispa_class_names)
print(leafblast_class_names)
```

### Contrast Adjustment (OPTIONAL)


There are photos of diseased and healthy paddies. To allow better recognition, we could tune up the contrast of the photo.

```python
CONTRAST_BOOST = 1.75    # [Multiplies element wise]
BRIGHTNESS_BOOST = 75 # [Remember on scale 0-255]
```

```python
def adjust_image(img):
    img = tf.clip_by_value( tf.image.adjust_brightness(img, BRIGHTNESS_BOOST), 0, 255 )
    return tf.clip_by_value( tf.image.adjust_contrast(img, CONTRAST_BOOST), 0, 255)
```

```python
# Training dataset contrast adjustment
train_brownspot_ds = train_brownspot_ds.map(lambda img, label: (adjust_image(img), label) )
train_hispa_ds = train_hispa_ds.map(lambda img, label: (adjust_image(img), label) )
train_leafblast_ds = train_leafblast_ds.map(lambda img, label: (adjust_image(img), label) )
```

```python
# Validation dataset contrast adjustment
validate_brownspot_ds = validate_brownspot_ds.map(lambda img, label: (adjust_image(img, CONTRAST_BOOST), label) )
validate_hispa_ds = validate_hispa_ds.map(lambda img, label: (adjust_image(img, CONTRAST_BOOST), label) )
validate_leafblast_ds = validate_leafblast_ds.map(lambda img, label: (adjust_image(img, CONTRAST_BOOST), label) )
```

### Caching

```python
# Dataset performance. Keep images in memory and prefetch for faster training
AUTOTUNE = tf.data.AUTOTUNE

train_brownspot_ds = train_brownspot_ds.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
train_hispa_ds = train_hispa_ds.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
train_leafblast_ds = train_leafblast_ds.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)

validate_brownspot_ds = validate_brownspot_ds.cache().prefetch(buffer_size=AUTOTUNE)
validate_hispa_ds = validate_hispa_ds.cache().prefetch(buffer_size=AUTOTUNE)
validate_leafblast_ds = validate_leafblast_ds.cache().prefetch(buffer_size=AUTOTUNE)
```

## 3.0 - Building Model

---
<br>


Each pixel values are between 0 to 255, which is not ideal for neural network. Scale them down into range [0,1].


For this, we can use a layer in our Sequential model, layers.experimental.preprocessing.Rescaling with scale of 1/255


Here I've loaded several pretrained models, but use them only if necessary

```python
IMG_SHAPE = (300, 300, 3)
```

```python
# Pretrained model - InceptionV3
inceptionV3_model = tf.keras.applications.inception_v3.InceptionV3(
    include_top=False, 
    weights='imagenet',
    input_shape=IMG_SHAPE
)

# Pretrained model - MobileNetV2
mobileNetV2_model = tf.keras.applications.MobileNetV2(input_shape=IMG_SHAPE,
                                                   include_top = False,
                                                   weights = 'imagenet')


# Pretrained model - ResNet50V2
resnet50v2_model = tf.keras.applications.ResNet50V2(
    include_top=False,
    weights="imagenet",
    input_shape=IMG_SHAPE
)

# Do not train pretrained model further - Takes a long time
inceptionV3_model.trainable = False
mobileNetV2_model.trainable = False
resnet50v2_model.trainable = False
```

### Data Augmentation Layer

```python
# Data Augmentation layers
# Flipping, Rotating
data_augmentation = keras.Sequential(
  [
    layers.experimental.preprocessing.RandomFlip("horizontal_and_vertical", input_shape=IMG_SHAPE),
    layers.experimental.preprocessing.RandomRotation(0.1),
  ]
)
```

### Models for 3 Diseases

```python
# The learning rate determines how much we're allowed to change the weights and biases
# of the model. Use small learning rate because we are using a already pre trained model
base_learning_rate = 0.001
```

```python
# Brownspot model
brownspot_model = keras.Sequential([
    data_augmentation,                                     # Data Augmentation
    layers.experimental.preprocessing.Rescaling(1./255),   # Scale [0,255] to [0,1]
    resnet50v2_model,                                      # Pretrained model
    layers.Flatten(), 
    layers.Dense(32, activation='relu'),
    layers.Dense(1, activation='sigmoid')                  # Healthy or Brown Spot
])

brownspot_model.compile(
    optimizer=tf.keras.optimizers.RMSprop(learning_rate=base_learning_rate),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

brownspot_model.summary()
```

```python
# Hispa model
hispa_model = keras.Sequential([
    data_augmentation,                                     # Data Augmentation
    layers.experimental.preprocessing.Rescaling(1./255),   # Scale [0,255] to [0,1]
    resnet50v2_model,                                     # Pretrained model
    layers.Flatten(), 
    layers.Dense(32, activation='relu'),
    layers.Dense(1, activation='sigmoid')                  # Healthy or Hispa
])

hispa_model.compile(
    optimizer=tf.keras.optimizers.RMSprop(learning_rate=base_learning_rate),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

hispa_model.summary()
```

```python
# Leaf Blast Model
leafblast_model = keras.Sequential([
    data_augmentation,                                     # Data Augmentation
    layers.experimental.preprocessing.Rescaling(1./255),   # Scale [0,255] to [0,1]
    resnet50v2_model,                                      # Pretrained model
    layers.Flatten(), 
    layers.Dense(32, activation='relu'),
    layers.Dense(1, activation='sigmoid')                  # Healthy or Leaf Blast
])

leafblast_model.compile(
    optimizer=tf.keras.optimizers.RMSprop(learning_rate=base_learning_rate),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

leafblast_model.summary()
```

```python
#### TESTING

test_hispa_model = keras.Sequential([
    data_augmentation,                                     # Data Augmentation
    layers.experimental.preprocessing.Rescaling(1./255),   # Scale [0,255] to [0,1]
    
    # Convolutional Layers
    layers.Conv2D(32, (5,5), activation='relu', input_shape=IMG_SHAPE),
    layers.MaxPool2D(2,2),
    layers.Dropout(0.2),
    
    layers.Conv2D(64, (5,5), activation='relu'),
    layers.MaxPool2D(2,2),
    layers.Dropout(0.2),
    
    layers.Conv2D(128, (5,5), activation='relu'),
    layers.MaxPool2D(2,2),
    layers.Dropout(0.2),
    
    # Flatten
    layers.Flatten(),
    
    # Dense layers for classification
    layers.Dense(512,activation='sigmoid'),
    
    layers.Dense(1,activation='sigmoid')
])

test_hispa_model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

test_hispa_model.summary()
```

```python
epochs = 10
history = test_hispa_model.fit(
  train_hispa_ds,
  validation_data=validate_hispa_ds,
  epochs=epochs
)
```

## 4.0 - Training The Models

```python
# Training the Brown Spot model
epochs = 10

brownspot_model_checkpoint = tf.keras.callbacks.ModelCheckpoint(
    filepath='trained_models/um_brownspot_model.h5',
    save_weights_only=False,
    monitor='val_accuracy',
    mode='max',
    save_best_only=True
)

history = brownspot_model.fit(
    train_brownspot_ds,
    validation_data=validate_brownspot_ds,
    epochs=epochs,
    callbacks=brownspot_model_checkpoint
)
```

```python
# Training the Leaf Blast Model
epochs = 10

leafblast_model_checkpoint = tf.keras.callbacks.ModelCheckpoint(
    filepath='trained_models/um_leafblast_model.h5',
    save_weights_only=False,
    monitor='val_accuracy',
    mode='max',
    save_best_only=True
)

history = leafblast_model.fit(
    train_hispa_ds,
    validation_data=validate_leafblast_ds,
    epochs=epochs,
    callbacks=leafblast_model_checkpoint
)
```
