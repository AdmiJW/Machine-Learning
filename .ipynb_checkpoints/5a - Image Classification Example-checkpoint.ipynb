# 5(a) - Image Classification Example

---
<br>

This is simply an hands-on example extracted from [Tensorflow's tutorial](https://www.tensorflow.org/tutorials/images/classification)

In this example, we will utilize `keras.Seqeuntial` model as well as loading data from `preprocessing.image_dataset_from_directory`

<br>

---

## 1.0 - Importing Modules

---
<br>

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
```

<br>

---

## 2.0 - Dataset

---
<br>

We'll make a flower classifier from 3670 photos, with classes of
* daisy
* dandelion
* roses
* sunflowers
* tulips

```python
import pathlib

# Download the photos and store in the filepath given
dataset_url = "https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz"
data_dir = tf.keras.utils.get_file( "flower_photos", origin=dataset_url, untar=True)
data_dir = pathlib.Path(data_dir)
```

```python
# Should have 3670 photos, in 5 subdirectories
print(len(list(data_dir.glob('*/*.jpg'))) )
```

### Loading Images using `keras.preprocessing`

```python
# Parameters for loading the images
batch_size = 32        # Batch size of 32 when training
img_height = 180       # Resize Height
img_width = 180        # Resize Width
```

```python
# Training Dataset
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
  data_dir,                 # Directory. Contains 5 subdirectory
  validation_split=0.2,     # 20% used for validation purpose. 80% use for testing
  subset="training",        # This is training dataset
  seed=123,                 # Seed for random shuffling and tranformation
  image_size=(img_height, img_width),   # Resize each image to this fixed size
  batch_size=batch_size                 # Batch size
)
```

```python
# Validation Dataset
val_ds = tf.keras.preprocessing.image_dataset_from_directory(
  data_dir,
  validation_split=0.2,
  subset="validation",
  seed=123,        # Use the same seed, or validation may have overlaps with training images (Unsure)
  image_size=(img_height, img_width),
  batch_size=batch_size
)
```

```python
# This is the class names of our photo. Should have 5 flower classes based on name of subdirectory
class_names = train_ds.class_names
print(class_names)
```

```python
# Let's see some flowers from our dataset
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))
# Take 1 batch (size 32), not 1 image
for images, labels in train_ds.take(1):
  for i in range(9):
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(images[i].numpy().astype("uint8"))     # Pass in as numpy array of type unsigned int 8 byte
    plt.title(class_names[labels[i]])                 # Label with correct class name
    plt.axis("off")
```

```python
# Obtain a image batch and label batch, check the shape of one batch
for image_batch, labels_batch in train_ds:
  print(image_batch.shape)                  # (32, 180, 180, 3) - 32 images, 180x180 size, 3 channels
  print(labels_batch.shape)                 # (32, ) - 32 labels
  break
```

```python
# Dataset performance. Keep images in memory and prefetch for faster training
AUTOTUNE = tf.data.AUTOTUNE

train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)
```

### Convert [0,255] to [0,1] 

Each pixel values are between 0 to 255, which is not ideal for neural network. Scale them down into range [0,1].
<br>

For this, we can use a layer in our Sequential model, `layers.experimental.preprocessing.Rescaling` with scale of 1/255

<br>

---

## 3.0 - Building Model

---
<br>

Let's try to use the pretrained model, `MobileNetV2`

```python
# Pretrained Model, MobileNetV2
IMG_SHAPE = (img_height, img_width, 3)
base_model = tf.keras.applications.MobileNetV2(input_shape=IMG_SHAPE,
                                               include_top = False,
                                               weights = 'imagenet')

base_model.trainable = False
```

```python
# Actual Model
num_classes = 5

model = Sequential([
  layers.experimental.preprocessing.Rescaling(1./255, input_shape=(img_height, img_width, 3)),
  base_model,
  layers.GlobalAveragePooling2D(),
  layers.Dense(num_classes)
])
```

```python
model.summary()
```

<br>

---

## 4.0 - Training the Model

---
<br>

```python
# The learning rate determines how much we're allowed to change the weights and biases
# of the model. Use small learning rate because we are using a already pre trained model
base_learning_rate = 0.001
```

```python
model.compile(
    optimizer=tf.keras.optimizers.RMSprop(learning_rate=base_learning_rate),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['accuracy']
)
```

```python
# Let's train our model
epochs=3
history = model.fit(
  train_ds,
  validation_data=val_ds,
  epochs=epochs
)
```

<br>

---

## 5.0 - Data Augmentation

---
<br>

Let's include some data augmentation in our datasets to further prevent overfitting. It can be done via several layers

### Flipping, Rotating, Zooming

```python
# Data Augmentation layers
data_augmentation = keras.Sequential(
  [
    layers.experimental.preprocessing.RandomFlip("horizontal", 
                                                 input_shape=(img_height, 
                                                              img_width,
                                                              3)),
    layers.experimental.preprocessing.RandomRotation(0.5),
    layers.experimental.preprocessing.RandomZoom(0.2),
  ]
)
```

```python
# Visualize data augmentation
plt.figure(figsize=(10, 10))
for images, _ in train_ds.take(1):
  for i in range(9):
    augmented_images = data_augmentation(images)
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(augmented_images[0].numpy().astype("uint8"))
    plt.axis("off")
```

### Dropout from Output Units

```python
# Included Data Augmentation and Dropouts (Actually not really need dropout due to pretrained model)
model = Sequential([
  data_augmentation,
  layers.experimental.preprocessing.Rescaling(1./255),
  base_model,
  layers.Dropout(0.2),
  layers.Flatten(),
  layers.Dense(num_classes)
])
```

```python
model.compile(
    optimizer=tf.keras.optimizers.RMSprop(learning_rate=base_learning_rate),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['accuracy']
)
```

```python
model.summary()
```

```python
# Train once again
epochs = 3
history = model.fit(
  train_ds,
  validation_data=val_ds,
  epochs=epochs
)
```

<br>

---

## 6.0 - Prediction

---
<br>

```python
sunflower_url = "https://storage.googleapis.com/download.tensorflow.org/example_images/592px-Red_sunflower.jpg"
sunflower_path = tf.keras.utils.get_file('Red_sunflower', origin=sunflower_url)

img = keras.preprocessing.image.load_img(
    sunflower_path, target_size=(img_height, img_width)
)
img_array = keras.preprocessing.image.img_to_array(img)
img_array = tf.expand_dims(img_array, 0) # Create a batch

predictions = model.predict(img_array)
score = tf.nn.softmax(predictions[0])

print(
    "This image most likely belongs to {} with a {:.2f} percent confidence."
    .format(class_names[np.argmax(score)], 100 * np.max(score))
)
```
